```r
#Libraries
library(arules)
library(arulesViz)
library(tidyverse)
```

```r
# Reading transactions
df <- read.transactions("../Dataset/Groceries.csv", sep=",")
```

```r
# View Transaction
inspect(df[1:10])
```

```r
# summarize whole data set
summary(df)
```

```r
# Frequency plot of top 15 items
itemFrequencyPlot(df, topN=15, type="absolute")
```

```r
# Get the rules
rules = apriori(df, parameter = list(supp = 0.003, conf = 0.3))
```

```r
# Summarize the rules
summary(rules)
```

```r
# Print the rules
inspect(rules[1:10])
```

```r
# Sort the rules by confidence
rules = sort(rules, by = "confidence")
options(digits = 2)
```

```r
inspect(rules[1:10])
```

```r
# Inspect the redundant rules
print("Showing first 10 rules")
rules[is.redundant(rules)]
inspect(rules[is.redundant(rules)][1:10])
rules = rules[!is.redundant(rules)]
```

```r
inspect(rules[1:10])
```

```r
# Plot the graphs for the rules
plot(rules, method = "graph")
```

```r
plot(rules, method = "paracoord")
```

```r
plot(rules, method = "matrix", control = list(reorder = "none"))
```
